```python
from db.connect_db import get_connection
import pandas as pd
from pathlib import Path

# Create connection
connection = get_connection()
cursor = connection.cursor()

print("Connected to StudyBuddy database.")
```

```python
# Path to the cleaned majors CSV
csv_path = Path("..") / "data" / "clean" / "us_majors.csv"

# Read the file
df_majors = pd.read_csv(csv_path)

print(f" Loaded {len(df_majors)} majors from us_majors.csv")
df_majors.head()
```

```python
insert_query = """
    INSERT INTO Majors (major_name)
    VALUES (%s)
"""

count = 0
for _, row in df_majors.iterrows():
    major_name = row['major_name'].strip()
    cursor.execute(insert_query, (major_name,))
    count += 1

# Commit all inserts
connection.commit()

print(f" Successfully inserted {count} majors into the Majors table!")
```
