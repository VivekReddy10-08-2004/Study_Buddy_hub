```python
import pandas as pd
from pathlib import Path

# Path to your colleges CSV file
csv_path = Path("..") / "data" / "clean" / "colleges_clean.csv"

# Load the CSV (only one column: college_name)
df_colleges = pd.read_csv(csv_path)

print(" Loaded colleges.csv successfully!")
print("Number of colleges:", len(df_colleges))
df_colleges.head()
```

```python
insert_query = """
    INSERT INTO Colleges (college_name)
    VALUES (%s)
"""

# Keep track of how many were inserted
count = 0

for _, row in df_colleges.iterrows():
    college_name = row['college_name'].strip()  # clean up whitespace
    cursor.execute(insert_query, (college_name,))
    count += 1

# Commit changes
connection.commit()

print(f" Successfully inserted {count} college records into the Colleges table!")
```
