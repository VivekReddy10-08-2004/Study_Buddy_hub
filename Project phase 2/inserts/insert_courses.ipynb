```python
import sys, os

# Tell Python to also look one folder up (where 'db' lives)
sys.path.append(os.path.abspath(".."))

# Optional: confirm it worked
print(" Project path added.")
```

```python
from db.connect_db import get_connection
import pandas as pd
from pathlib import Path

# Create database connection
connection = get_connection()
cursor = connection.cursor()

print(" Connected to StudyBuddy database.")
```

```python
# Path to the cleaned USM courses file
csv_path = Path("..") / "data" / "clean" / "usm_courses_clean.csv"

# Load into pandas
df_courses = pd.read_csv(csv_path)

print(f" Loaded {len(df_courses)} rows from usm_courses_clean.csv")
df_courses.head()
```

```python
insert_query = """
    INSERT INTO Courses (course_code, course_name, college_id)
    VALUES (%s, %s, %s)
"""

count = 0
for _, row in df_courses.iterrows():
    course_code = str(row["course_code"]).strip()
    course_name = str(row["course_name"]).strip()
    college_id = int(row["college_id"])
    
    cursor.execute(insert_query, (course_code, course_name, college_id))
    count += 1

connection.commit()

print(f" Successfully inserted {count} courses into the Courses table!")
```

```python
cursor.execute("SELECT COUNT(*) FROM Courses;")
print(" Total courses in table:", cursor.fetchone()[0])

cursor.execute("SELECT course_id, course_code, course_name, college_id FROM Courses LIMIT 10;")
for row in cursor.fetchall():
    print(row)
```

```python
cursor.close()
connection.close()
print(" Connection closed.")
```
